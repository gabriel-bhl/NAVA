```python
import pandas as pd
import numpy as np

# Drop rows where manufacturer or model has more than 35 characters
dados_processados = dados_processados[dados_processados['manufacturer'].map(len) <= 30]
dados_processados = dados_processados[dados_processados['model'].map(len) <= 30]

# Drop rows where year, price or odometer are not numbers
# First, convert the columns to numeric, errors='coerce' will replace non-numeric values with NaN
dados_processados['year'] = pd.to_numeric(dados_processados['year'], errors='coerce')
dados_processados['price'] = pd.to_numeric(dados_processados['price'], errors='coerce')
dados_processados['odometer'] = pd.to_numeric(dados_processados['odometer'], errors='coerce')

# Then drop the rows with NaN values in these columns
dados_processados = dados_processados.dropna(subset=['year', 'price', 'odometer'])
```

Normalizando as colunas numéricas

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

dados_processados_sem_dupla[col_num] = scaler.fit_transform(dados_processados_sem_dupla[col_num])
```

Método "Elbow" para descobrir o melhor número de clusters

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

K = range(1,11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0).fit(dados_processados_sem_dupla[col_num])
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# First, inverse transform the standardized values back to their original scale
original_values = scaler.inverse_transform(dados_processados_sem_dupla[col_num])

# Convert this array back to a DataFrame
df_original_values = pd.DataFrame(original_values, columns=col_num, index=dados_processados_sem_dupla.index)

# Add the cluster labels
df_original_values['cluster'] = kmeans.labels_

# Now you can plot the scatter plot with original values
plt.figure(figsize=(10, 8))

# Scatter plot of price vs odometer colored by cluster assignment
plt.scatter(df_original_values['odometer'], df_original_values['price'], c=df_original_values['cluster'], cmap='viridis')

plt.xlabel('Milhagem')
plt.ylabel('Preço')
plt.title('Clusters Milhagem vs Preço)')
plt.show()
```
